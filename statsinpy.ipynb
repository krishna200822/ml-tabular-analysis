```python
import random
import time
import numpy as np
from collections import Counter
from scipy import stats
```

```python
# Genrate 10,00,000 random integer bw 1 to 1000
data =[random.randint(1,1000) for _  in range(1000000)]
np_data =np.array(data)
```

```python
# 1 From Scratch Implementation
def mean_from_scratch(data):
    return sum(data)/len(data)
def median_from_scratch(data):
    sorted_data = sorted(data)
    n = len(sorted_data)
    mid =n// 2
    if n % 2 == 0:
      return (sorted_data[mid - 1] + sorted_data[mid]) / 2
    else:
      return sorted_data[mid]

def mode_from_scratch(data):
    freq = Counter(data)
    max_freq = max(freq.values())
    mode = [k for k, v in freq.items() if v == max_freq]
    return mode[0] if len(mode) == 1 else mode
```

```python

# timing scratch
start = time.time()
mean_s= mean_from_scratch(data)
medion_s = median_from_scratch(data)
mode_s = mode_from_scratch(data)
end = time.time()
scratch_time = end - start
```

```python
# numpy built in implemntation
start = time.time()
mean_np= np.mean(np_data)
medion_np = np.median(np_data)
mode_np = stats.mode(np_data,keepdims=True).mode
end = time.time()
numoy_time = end - start
```

```python
# results
print("From scratch:")
print("Mean:", mean_s)
print("Median:", medion_s)
print("Mode:", mode_s)
print("Time taken:", scratch_time)

print("\nUsing numpy:")
print("Mean:", mean_np)
print("Median:", medion_np)
print("Mode:", mode_np)
print("Time taken:", numoy_time)
```
