```python
class Node:
  def __init__(self,data,):
    self.data=data
    self.next=None
a=Node(5)# assigning nodes
b=Node(3)
c=Node(7)
a.next=b # connecting nodes
b.next =c
head=a

print(head.data)
print(head.next.data)
```

```python
#traverse (to print all link list)
def printLInkedList(head):
    curr=head
    while curr is not None:
        print(curr.data)
        curr=curr.next
printLInkedList(head)
```

```python
# insertion at begning
new_node=Node(4)
new_node.next=head
head=new_node
```

```python
# insertion at end
new_node=Node(1)
curr=head
while curr.next is not None:
  curr=curr.next
printLInkedList(head)

```

```python
# insertion at kth index
k=2
curr=head
for i in range (k-1):
  curr = curr.next
  new_node=Node(6)
  new_node.next=curr.next
  curr.next=new_node
printLInkedList(head)
```

```python
# deletion at kth index
k=2
curr=head
for i in range (k-1):
  curr=curr.next
curr.next=curr.next.next
printLInkedList(head)

```

```python
class LinkedList:
    def __init__(self):
        self.head = None

    def insert_at_beginning(self, data):
        new_node = Node(data)
        new_node.next = self.head
        self.head = new_node

    def insert_at_end(self, data):
        new_node = Node(data)
        if self.head is None:
            self.head = new_node
            return
        curr = self.head
        while curr.next:
            curr = curr.next
        curr.next = new_node

    def insert_at_kth_index(self, data, k):
        if k < 0:
            print("Invalid index.")
            return
        if k == 0:
            self.insert_at_beginning(data)
            return

        new_node = Node(data)
        curr = self.head
        for _ in range(k - 1):
            if curr is None:
                print("Index out of bounds.")
                return
            curr = curr.next

        if curr is None:
            print("Index out of bounds.")
            return

        new_node.next = curr.next
        curr.next = new_node

    def delete_at_kth_index(self, k):
        if self.head is None:
            print("List is empty. Cannot delete.")
            return
        if k < 0:
            print("Invalid index.")
            return

        if k == 0:
            self.head = self.head.next
            return

        curr = self.head
        for _ in range(k - 1):
            if curr.next is None:
                print("Index out of bounds.")
                return
            curr = curr.next

        if curr.next is None:
            print("Index out of bounds.")
            return

        curr.next = curr.next.next

    def search_node(self, data):
        curr = self.head
        position = 0
        while curr:
            if curr.data == data:
                return position
            curr = curr.next
            position += 1
        return -1 # Not found

    def display_list(self):
        curr = self.head
        if curr is None:
            print("Linked List is empty.")
            return
        elements = []
        while curr:
            elements.append(str(curr.data))
            curr = curr.next
        print(" -> ".join(elements))

    def run_menu(self):
        print("\n--- Linked List Menu ---")
        print("1. Insert node at beginning")
        print("2. Insert node at end")
        print("3. Insert node at k-th index")
        print("4. Delete node at k-th index")
        print("5. Search node")
        print("6. Display list")
        print("7. Exit")
```
